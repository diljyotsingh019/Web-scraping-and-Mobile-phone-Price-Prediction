```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd
```

```python
path = "C:\Program Files (x86)\chromedriver.exe"
driver = webdriver.Chrome(path)
driver.get("https://amazon.in")
```

```python
def search(product_name):
    try:
        element = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID,"twotabsearchtextbox")))
        element.send_keys(product_name)
        element.send_keys(Keys.RETURN)
    except:
        driver.quit()
```

```python
search("Mobile Phones")
```

```python
final_list = []
for i in range(20):
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR,"span.a-size-medium.a-color-base.a-text-normal")))
    count = driver.find_elements_by_css_selector("span.a-size-medium.a-color-base.a-text-normal")
    for j in range(len(count)):
        WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR,"span.a-size-medium.a-color-base.a-text-normal")))
        driver.find_elements_by_css_selector("span.a-size-medium.a-color-base.a-text-normal")[j].click()
        driver.switch_to.window(driver.window_handles[1])
        l = []
        features = driver.find_elements_by_class_name("base-item-column")
        l.append(driver.find_element_by_id("productTitle").text)
        try: 
            l.append((driver.find_element_by_id("priceblock_ourprice").text))
        except:
            try:
                l.append((driver.find_element_by_id("priceblock_dealprice").text))
            except:
                l.append("")
        for i in features:
            l.append(i.text)
        final_list.append(l[0:13])
        driver.close()
        driver.switch_to.window(driver.window_handles[0])
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR,"li.a-last"))).click()
```

```python
pd.DataFrame(final_list)
```

```python
pd.DataFrame(final_list).to_csv("AmazonMobilePhones.csv")
```
